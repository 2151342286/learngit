# git的使用

## 1.创建库和添加文件
### 1.1 初始化目录为git可以管理的仓库
进入目标目录，输入`git init`  
之后目录下会多一个.git目录，用于跟踪管理版本库，看不到是因为被隐藏，用`ls -ah`可以看见  
### 1.2 添加文件
1. 在项目目录内编写文件
2. 使用`git add 文件`将文件添加到仓库
3. 使用` git commit -m "xxx"`将文件提交，`-m`可以不写，但不建议
4. `commit`可以一次提交很多文件
### 1.3 常见问题
1. 使用Windows系统请确保目录名（包括父目录）不包含中文。
2. Git命令必须在Git仓库目录内执行（`git init`除外），在仓库目录外执行是没有意义的。 
3. 添加某个文件时，该文件必须在当前目录下存在

## 2.文件管理
### 2.1 工作区状态
1. 修改文件后使用`git status`查看仓库状态，可见修改情况
    ```bash
    $ git status
    On branch master
    Changes not staged for commit:
    (use "git add <file>..." to update what will be committed)
    (use "git restore <file>..." to discard changes in working directory)
            modified:   learn.ipynb

    no changes added to commit (use "git add" and/or "git commit -a")
    ```
2. 看改变详情用`git diff`,如`git diff learn.ipynb`(按q退出)
### 2.2 版本回退
1. 使用`git log`查看历史记录，加`--pretty=oneline`可简化输出
2. 一大串类似`1094adb...`的是`commit id`（版本号）
3. `HEAD`表示当前版本，上一个版本就是`HEAD^`,上上个就是`HEAD^^`,上100个可写作`HEAD~100`
4. 使用`git reset commit_id`到达某个版本，例如`git reset --hard HEAD^`就是回退到上一个版本
5. `git reset` 的三种模式：
   - `--hard`：回退到上个版本的已提交状态
   - `--soft`：回退到这个版本的已添加未提交状态
   - `--mixed`：回退到这个版本未添加状态
6. 如果错误回退想回到最新版本,可以使用`git reset --hard xxxx`,`xxxx`是想返回的版本的版本号,适当写前几位就行
7. 查看版本号可以用`git reflog`,使用它可以看见命令记录
8. **特别注意,如果`reset`且使用`--hard`,要记得先提交目前工作区内容，否则会丢失！！！**


### 2.3 工作区和暂存区
- 工作区（Working Directory）：电脑里能看到的目录
- 版本库（Repository）：工作区里的一个隐藏目录`.git`，不算工作区,，内含暂存区，分支和`HEAD`
- 暂存区（stage/index）：`git add`把文件添加进去，实际上就是把文件修改添加到暂存区
- 分支:Git为我们自动创建的第一个分支`master`,指向`master`的一个指针叫`HEAD`

### 2.4 管理修改
只有使用add的修改才可以被commit，例如：修改文件x->`git add x`->再次修改x->git commit,此时只有第一次修改会被提交

### 2.5 撤销修改
1. 丢弃工作区修改  
`git checkout -- readme.txt`：把`readme.txt`文件在工作区的修改全部撤销，这里有两种情况：  
- 一种是`readme.txt`自修改后还没有被放到暂存区，现在，撤销修改就回到和版本库一模一样的状态。  
- 一种是`readme.txt`已经添加到暂存区后，又作了修改，现在，撤销修改就回到添加到暂存区后的状态。  
- 就是让这个文件回到最近一次git commit或git add时的状态，即用暂存区/版本库覆盖工作区。  
2. 撤销暂存区  
`git reset HEAD <file>`:可以把暂存区的修改撤销掉（unstage），重新放回工作区  
例如：`git reset HEAD learn.ipynb`

3. 以上为旧指令,可使用但不推荐，以下是同功能新指令
- git switch 切换分支
    ```bash
    git switch <分支名>        # 切换分支
    git switch -c <新分支名>    # 创建并切换到新分支
    ```
- git restore 撤销修改
    ```bash 
    git restore <文件>              # 丢弃工作区修改（用暂存区覆盖）
    git restore --staged <文件>     # 撤销暂存（unstage）
    git restore --source=HEAD <文件> # 用版本库覆盖工作区
    ```

### 2.6 删除文件
1. 本地删除（之后用`git status`可见删除情况）
- 确定删除：使用`git rm 文件名`删除文件，再`commit`一下  
- 误删解决：使用`git checkout -- 文件名`可恢复文件（用版本库里的版本替换工作区的版本，可参考上一小节）
