# git的使用

## 1.创建库和添加文件
### 1.1 初始化目录为git可以管理的仓库
进入目标目录，输入`git init`  
之后目录下会多一个.git目录，用于跟踪管理版本库，看不到是因为被隐藏，用`ls -ah`可以看见  
### 1.2 添加文件
1. 在项目目录内编写文件
2. 使用`git add 文件`将文件添加到仓库
3. 使用` git commit -m "xxx"`将文件提交，`-m`可以不写，但不建议
4. `commit`可以一次提交很多文件
5. `git add .`:把当前目录下的所有修改（包括新增、删除、修改）添加到暂存区。
### 1.3 常见问题
1. 使用Windows系统请确保目录名（包括父目录）不包含中文。
2. Git命令必须在Git仓库目录内执行（`git init`除外），在仓库目录外执行是没有意义的。 
3. 添加某个文件时，该文件必须在当前目录下存在

## 2.文件管理
### 2.1 工作区状态
1. 修改文件后使用`git status`查看仓库状态，可见修改情况
    ```bash
    $ git status
    On branch master
    Changes not staged for commit:
    (use "git add <file>..." to update what will be committed)
    (use "git restore <file>..." to discard changes in working directory)
            modified:   learn.ipynb

    no changes added to commit (use "git add" and/or "git commit -a")
    ```
2. 看改变详情用`git diff`,如`git diff learn.ipynb`(按q退出)
### 2.2 版本回退
1. 使用`git log`查看历史记录，加`--pretty=oneline`可简化输出,加`--graph`以图方式显示
2. 一大串类似`1094adb...`的是`commit id`（版本号）
3. `HEAD`表示当前版本，上一个版本就是`HEAD^`,上上个就是`HEAD^^`,上100个可写作`HEAD~100`
4. 使用`git reset commit_id`到达某个版本，例如`git reset --hard HEAD^`就是回退到上一个版本
5. `git reset` 的三种模式：
   - `--hard`：回退到上个版本的已提交状态
   - `--soft`：回退到这个版本的已添加未提交状态
   - `--mixed`：回退到这个版本未添加状态
6. 如果错误回退想回到最新版本,可以使用`git reset --hard xxxx`,`xxxx`是想返回的版本的版本号,适当写前几位就行
7. 查看版本号可以用`git reflog`,使用它可以看见命令记录
8. **特别注意,如果`reset`且使用`--hard`,要记得先提交目前工作区内容，否则会丢失！！！**


### 2.3 工作区和暂存区
- 工作区（Working Directory）：电脑里能看到的目录
- 版本库（Repository）：工作区里的一个隐藏目录`.git`，不算工作区,，内含暂存区，分支和`HEAD`
- 暂存区（stage/index）：`git add`把文件添加进去，实际上就是把文件修改添加到暂存区
- 分支:Git为我们自动创建的第一个分支`master`,指向`master`的一个指针叫`HEAD`

### 2.4 管理修改
只有使用add的修改才可以被commit，例如：修改文件x->`git add x`->再次修改x->git commit,此时只有第一次修改会被提交

### 2.5 撤销修改
1. 丢弃工作区修改  
`git checkout -- readme.txt`：把`readme.txt`文件在工作区的修改全部撤销，这里有两种情况：  
- 一种是`readme.txt`自修改后还没有被放到暂存区，现在，撤销修改就回到和版本库一模一样的状态。  
- 一种是`readme.txt`已经添加到暂存区后，又作了修改，现在，撤销修改就回到添加到暂存区后的状态。  
- 就是让这个文件回到最近一次git commit或git add时的状态，即用暂存区/版本库覆盖工作区。  
2. 撤销暂存区  
`git reset HEAD <file>`:可以把暂存区的修改撤销掉（unstage），重新放回工作区  
例如：`git reset HEAD learn.ipynb`

3. 以上为旧指令,可使用但不推荐，以下是同功能新指令
- git switch 切换分支
    ```bash
    git switch <分支名>        # 切换分支
    git switch -c <新分支名>    # 创建并切换到新分支
    ```
- git restore 撤销修改
    ```bash 
    git restore <文件>              # 丢弃工作区修改（用暂存区覆盖）
    git restore --staged <文件>     # 撤销暂存（unstage）
    git restore --source=HEAD <文件> # 用版本库覆盖工作区
    ```

### 2.6 删除文件
1. 本地删除（之后用`git status`可见删除情况）
- 确定删除：使用`git rm 文件名`删除文件，再`commit`一下  
- 误删解决：使用`git checkout -- 文件名`可恢复文件（用版本库里的版本替换工作区的版本，可参考上一小节）


## 3.远程仓库
### 3.1 创建shh key
1. 在`git bash`中输入`ssh-keygen -t rsa -C "youremail@example.com"`
2. 在用户目录中找到公钥`id_rsa.pub`
3. 在GitHub上添加shh key,title随意,key填公钥内容
### 3.2 添加远程库
1. 登录github新建库

2. 在本地仓库下执行`git remote add origin git@github.com:你的github用户名/learngit.git`
- 添加后，远程库的名字就是origin，这是Git默认的叫法
3. 把本地库的所有内容推送到远程库上`git push -u origin master`
- 由于远程库是空的，我们第一次推送master分支时，加上了-u参数，Git不但会把本地的master分支内容推送的远程新的master分支，还会把本地的master分支和远程的master分支关联起来，在以后的推送或者拉取时就可以简化命令。
4. 之后本地提交，使用`git push origin master`就可推送到github  

5. 删除和查看远程库
- git remote rm <name> ：删除远程库（仅是解绑）
- git remote -v查看远程库信息
### 3.3 克隆远程库
git clone 库地址  
### **补充：这部分大多数操作要关闭加速器如steam++**

## 4.分支管理
### 4.1 创建与合并分支
1. `git checkout -b dev`:加上`-b`表示创建并切换到`dev`分支,等效于下面两条命令
- `git branch dev`：创建分支dev
- `git checkout dev`：转到分支dev
2. `git branch`:查看当前分支

3. `git merge dev`:将dev分支合并到当前分支
- 例如在dev上编写内容并提交后，转到master分支，然后再执行合并，master分支得到dev分支内容
- `Fast-forward`：“快进模式”，也就是直接把master指向dev的当前提交，所以合并速度非常快。
4. `git branch -d dev`:删除dev分支

5. 切换分支有新方法，在2.5节中提过,即：  
- `git switch -c dev`:创建并切换到新的dev分支
- `git switch master`:直接切换到已有的master分支
### 4.2 解决冲突
1. 创建一个分支dev，转到dev修改内容并add，再回到master，修改同一处内容并add，此时合并显示有冲突，因为改动重合，git无法判断要之后commit哪一个，必需解决冲突才能合并
- 冲突原因不唯一，比如两个分支都新建文件也会冲突
### 4.3 分支管理策略
1. 通常，合并分支时，如果可能，Git会用`Fast forward`模式，但这种模式下，删除分支后，会丢掉分支信息。
2. 强制禁用`Fast forward`模式，Git就会在`merge`时生成一个新的`commit`，这样，从分支历史上就可以看出分支信息。
3. `git merge --no-ff -m "merge with no-ff" dev`:合并dev到当前分支，且禁用`Fast forward`模式
### 4.4 stash临时保存功能
1. 场景：修改内容并`add`后要创建新分支修改一个bug,但现分支任务未完成，不想`commit`
2. 工作流程
    ```bash
    # 1. 查看当前状态
    git status
    # On branch master
    # Changes not staged for commit:
    #   modified: learn.ipynb

    # 2. 临时保存
    git stash
    # Saved working directory and index state WIP on master: ...

    # 3. 工作区变干净了
    git status
    # nothing to commit, working tree clean

    # 4. 做其他事情（切换分支、拉取代码等）
    git switch other

    # 5. 恢复之前的工作
    git stash pop
    # 修改又回来了
    ```
3. 部分指令功能
- `git stash`:把当前修改藏起来，让工作区变干净
- `git stash pop`:把最后一次存的修改恢复，并删除对应stash
- `git stash list`：列出所有stash
- 用`git stash push -m "你的备注"`保存,`git stash list`就能看到备注
- **注意，使用这个方法会把暂存区内容保存下来，同时，暂存区内容丢失，恢复时也只有工作区有原来内容，需要重新添加到暂存区**
### 4.5 强制删除分支
分支未合并删除会失败,将`-d`改为`-D`强制删除,如：`git branch -D feature`;  
只有分支有commit过内容，要强制删除才要`-D`
### 4.6 多人协作
1. `git remote`:查看远程库的信息,加`-v`显示更详细信息,，即`git remote -v`

2. `git push origin master`:推送分支，把master上的所有本地提交推送到远程库

3. 抓取分支
- git clone后，只能看见本地的master分支，要在dev分支上开发，就必须创建远程origin的dev分支到本地，即使用`git checkout -b dev origin/dev`
- git pull上传时可能会有冲突
4. `git rebase`:让提交历史变成一条直线，看起来更干净。但只能在个人分支上用，公共分支不能用！

## 5.标签管理
### 5.1 创建标签
1. 切换到想打标签的分支，然后输入`git tag <name>`即可，用`git tag`可以查看所有标签

2. 默认标签是打在最新提交的`commit`上的,用`git log --pretty=oneline --abbrev-commit`
可以查看历史提交`commit_id`

3. 想给其他其他`commit`打标签，查看完`commit_id`后在原来基础上加上其`id`。例如：`git tag newtag f17c996`

4. 使用`git tag`查看标签时标签按字母排序

5. 可以创建带有说明的标签，用-a指定标签名，-m指定说明文字：`git tag -a v0.1 -m "version 0.1 released" 1094adb`

6. 命令`git show <tagname>`可以看到说明文字
### 5.2 标签操作
1. `git tag -d <tagname>`:删除标签
- 因为创建的标签都只存储在本地，不会自动推送到远程。所以，打错的标签可以在本地安全删除。

2. 如果要推送某个标签到远程，使用命令`git push origin <tagname>`

3. 一次性推送全部尚未推送到远程的本地标签：`git push origin --tags`

4. 如果标签已经推送到远程，要删除远程标签
- 先删除本地标签` git tag -d  <tagname>`
- 然后，从远程删除。删除命令也是`push`，格式为`git push origin :refs/tags/ <tagname>`
